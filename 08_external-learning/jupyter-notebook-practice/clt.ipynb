```python
%matplotlib inline

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

random.seed(123)

sequences = []
all_values = []

for i in range(10000):
    sequence = []
    for j in range(100):
        x = random.expovariate(1)
        sequence.append(x)
        all_values.append(x)
    sequences.append(sequence)

np_sequences = pd.DataFrame(np.transpose(np.array(sequences)))

means = np_sequences.apply(np.mean)

plt.hist(np.array(all_values), bins = 100, histtype='step')
plt.show()
plt.clf()

plt.hist(means, bins = 20)
plt.show()
plt.clf()
```
